```python
cases = pd.read_csv('./data/covid_confirmed_usafacts.csv', parse_dates=True, infer_datetime_format=True, low_memory=False).convert_dtypes().set_index('countyFIPS').iloc[1:,3:]
deaths = pd.read_csv('./data/covid_deaths_usafacts.csv', parse_dates=True, infer_datetime_format=True, low_memory=False).convert_dtypes().set_index('countyFIPS').iloc[1:,3:]
pop =  pd.read_csv('./data/covid_county_population_usafacts.csv', parse_dates=True, infer_datetime_format=True, low_memory=False).convert_dtypes().set_index('countyFIPS')
nyt_counties = pd.read_csv('https://github.com/nytimes/covid-19-data/raw/master/us-counties.csv', parse_dates=True, infer_datetime_format=True, low_memory=False).convert_dtypes()
nyt_counties['date'] = pd.to_datetime(nyt_counties.date)

rt = pd.read_csv('https://d14wlfuexuxgcm.cloudfront.net/covid/rt.csv', parse_dates=True, infer_datetime_format=True, low_memory=False).convert_dtypes()
rt['date'] = pd.to_datetime(rt.date)

nyt_states = pd.read_csv('https://github.com/nytimes/covid-19-data/raw/master/us-states.csv', parse_dates=True, infer_datetime_format=True, low_memory=False).convert_dtypes()
nyt_states['date'] = pd.to_datetime(nyt_states.date)

county_cases = nyt_counties.pivot_table('cases', 'fips', 'date')
county_deaths = nyt_counties.pivot_table('deaths', 'fips', 'date')

state_cases = nyt_states.pivot_table('cases', 'fips', 'date')
state_deaths = nyt_states.pivot_table('deaths', 'fips', 'date')
```

```python
excluded_states = ['AK', 'HI', 'PR','VI', 'MP', 'AS', 'GU']

state_geo_src = 'https://raw.githubusercontent.com/loganpowell/census-geojson/master/GeoJSON/20m/2019/state.json'
county_geo_src = 'https://github.com/loganpowell/census-geojson/raw/master/GeoJSON/20m/2019/county.json'

#@memoize
def read_geo(path):
        return gpd.read_file(path)#.convert_dtypes()

county_geo = read_geo(county_geo_src)
state_geo = read_geo(state_geo_src)

df = county_geo.merge(state_geo, on='STATEFP', suffixes=['', '_state'])
df = df[~df.STUSPS.isin(excluded_states)]
df = df.set_index(df.GEOID.astype(int))

#gpl.choropleth(df,hue=log2(0.5+df.join(cases, how='inner').iloc[:, -1]), extent=df.total_bounds, projection=gpl.crs.AlbersEqualArea())
#gpl.choropleth(county_geo, hue=log2(county_geo.ALAND), extent=state_geo[~state_geo.STUSPS.isin(excluded_states)].total_bounds)
#gpl.choropleth(df,hue=log2(0.5+cases.loc[df.index].iloc[:, -1]), extent=df.total_bounds, projection=gpl.crs.AlbersEqualArea())
```
